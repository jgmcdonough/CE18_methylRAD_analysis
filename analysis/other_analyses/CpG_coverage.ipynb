create a function that takes in two dfs, and pulls out a subset that matches the genomic feature

```r
subset_CpGs <- function(df1, df2) {
    # find the genomic range of the first df (the CpGs)
    df1_gr <- GRanges(seqnames = df1$chromosome, 
                  ranges = IRanges(start = df1$start, end = df1$stop))
    
    # find the genomic range of the second df (genomic feature)
    df2_gr <- GRanges(seqnames = df2$chromosome, 
                  ranges = IRanges(start = df2$start, end = df2$stop))

    # find overlaps
    overlaps <- findOverlaps(df1_gr, df2_gr, type = "within")

    # get the subset of cpgs within the genomic feature
    cpg_overlap <- df1[queryHits(overlaps), ]
    
    return(cpg_overlap)
    }
```

```r
cpg_exons <- subset_CpGs(merged_df, exon)
cpg_introns <- subset_CpGs(merged_df, intron)
cpg_utrs <- subset_CpGs(merged_df, utr)
cpg_genes <- subset_CpGs(merged_df, gene)
```

need to select how I'm going to classify each group

from looking at the jitter plots, starting with this classification:
- **highly methylated**: 151-200+
- **moderately methylated**: 101-150
- **sparsely methylated**: 51-100
- **not methylated**: 0-50

```r
head(zero_inDMG)
```

```r
dmg_grouping <- zero_inDMG %>% 
  mutate(methyl_group = case_when(
    count <= 50 ~ "not methylated",
    count <= 100 ~ "sparsely methylated",
    count <= 150 ~ "moderately methylated",
    count >= 151 ~ "highly methylated"
  ))

head(dmg_grouping)
```

```r
filter(zero_inDMG, zero_inDMG$count > 150)
```

```r
options(repr.plot.width=10, repr.plot.height=15)

# reorder features and methyl group  
dmg_grouping$feature <- fct_relevel(dmg_grouping$feature, "gene", "exon", "intron", "utr")
dmg_grouping$methyl_group <- fct_relevel(dmg_grouping$methyl_group, 
                                        "not methylated", "sparsely methylated", 
                                        "moderately methylated", "highly methylated")

ggplot(dmg_grouping, aes(x=treatment, y=count)) +
geom_jitter(col = 'gray') +
geom_violin() + 
geom_boxplot(width=0.1) +
facet_grid(feature ~ methyl_group) +
theme_bw()
```

```r
prop_methyl <- dmg_grouping %>%
group_by(feature) %>%
count(methyl_group) %>%
group_by(feature) %>%
  mutate(proportion = n / sum(n) * 100)

head(prop_methyl)
```
